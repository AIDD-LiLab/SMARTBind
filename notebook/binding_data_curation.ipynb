## Build pickle file for SMARTBind model training

```python
import pickle
with open('../training_data/hariboss_merged_10fd.pkl', 'rb') as f:
    data = pickle.load(f)
```

The python dictionary structure should be as follows:
- `data.keys()`: RNA-ligand complex id;
- `data['3ds7'].keys()`: id for specific RNA-ligand pairs in a single PDB, the different formating of this id name will not affect the processing pipeline.

Here is example:

```python
list(data.keys())[0:5]
```

```python
data['3ds7'].keys()
```

```python
data['3ds7']['3ds7_B_Q_GNG_320']['train_split']
```

The minimum information requirements for `data['3ds7']['3ds7_B_Q_GNG_320'].keys()` (based on example above):
- 'pdb_id': '3ds7'
- 'train_split': [True, False, True]  --> this key is used for fold-splitting. If the `fold_num` in `binding_training.yaml` file is set to 1, then the pairs will be assigned into training dataset if the first element in the `train_split`  is True, otherwise it will be assigned into the validation dataset.
- 'rna_chain_id': 'B'
- 'rna_chain_sequence': 'GGACAUACAAUCGCGUGGAUAUGGCACGCAAGAUCCCGCCGGGCACCGUAAAUGUCCGACUAUGUCC' 
- 'downloaded_ligand_smiles': 'Nc1nc(=O)c2ncn([C@H]3C[C@H](O)[C@@H](CO)O3)c2[nH]1'
- 'download_fp2': [0, 0, 0, 0, 0, ..., 1, 0] --> 1024-dimension list
- 'ligand_id': 'GNG'
- 'contact_map_1d': [0, 1, 1, 0, 0, ..., 1, 0] --> binary groundtruth of the binding site in list, the length should be equal to the length of RNA chain sequence.
- 'decoy_smiles': augmented SMILES sampled by ligand-specific decoy augmentation strategy.

When the data construction is finished, you can save the pickle file back to `../training_data` folder, then specify the data path in the yaml configuration for SMARTBind model training.
