# Build decoy augmentation training data

Download decoy library using `gdown --id 1gU688NQjQG3zccePCZ0H_jbK2VJF66UT -O decoy_library.smi`

```python
import sys
sys.path.append("..") 
import pickle
from smartbind.preprocess import convert_smiles_to_pf2
from smartbind.decoy import top_decoys
import tqdm
```

```python
# read the smi file
with open('decoy_library.smi', 'r') as f:
    decoy_smiles = [line.strip().split()[0] for line in f.readlines()]
print(f"Total decoys loaded: {len(decoy_smiles)}")

decoy_smiles_fp2 = [convert_smiles_to_pf2(smi) for smi in decoy_smiles]
```

```python
with open('../training_data/hariboss_merged_10fd.pkl', 'rb') as f:
    training_data = pickle.load(f)
print(f"Total training PDBs loaded: {len(training_data)}")
```

```python
for k, v in tqdm.tqdm(training_data.items()):
    for kk, vv in v.items():
        fp2_str = ''.join(map(str, vv['download_fp2']))
        top_similar_fp = top_decoys(decoy_smiles_fp2, top_num=200, anchor_fp=fp2_str, tanimoto_threshold=0.5)
        vv['decoy_smiles'] = top_similar_fp
```
